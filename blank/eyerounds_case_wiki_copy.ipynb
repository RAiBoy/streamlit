```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://eyerounds.org/cases.htm'
resp=requests.get(url)
```

```python
soup = BeautifulSoup(resp.text, 'html.parser')
```

```python
class_list = soup.find_all(class_ = 'col-xs-12 col-sm-6 col-md-6 col-lg-4')
```

```python
link_list = []

for i in range(len(class_list)):
    class_a = class_list[i].find_all('a')
    for link in class_a:
        temp = link.get('href')
        start = temp.startswith('http://webeye.ophth.uiowa.edu/eyeforum/')
        if start == False:
            temp = 'http://webeye.ophth.uiowa.edu/eyeforum/'+temp
#         print(temp)  
        if temp.count('/')==5:
            link_list.append(temp)
            print(temp)
        else: continue    
```

```python
case_list = []
```

```python
for link in link_list:
    text_list = []
    case_resp = requests.get(link)
    case_soup = BeautifulSoup(case_resp.text, 'html.parser')
    temp = case_soup.select('div[class="col-md-12 mt-4 text-left"]')
#     print(temp)
#     temp_list.append(temp)
    
    for i in temp:
        text = i.get_text()
        text_list.append(text)
    case_list.append(text_list)    
```

```python
from medcat.cat import CAT
cat = CAT.load_model_pack('/Users/congwang/Downloads/medmen_wstatus_2021_oct')

```

```python
med_term_list = []
index = 0
for case in case_list:
    if case != []:
#         print(index)
        text = case[0]
        index +=1
        entities = cat.get_entities(text)
        med_terms = []
        for key in entities['entities']:
            med_terms.append(entities['entities'][key]['pretty_name'])
        med_term_list.append(med_terms)    
    else: continue 
```

```python
import re
from wikiapi import WikiApi
wiki = WikiApi()
wiki = WikiApi({ 'locale' : 'en'})

find = re.compile(r"^([^.]*).*")

wiki_list = []

explanation_list = []

for med_terms in med_term_list:
    wiki_dict = {}
    
    for i in range(len(med_terms)):
        results = wiki.find(med_terms[i])
        if len(results) == 0:
            continue
        else:
            article = wiki.get_article(results[0])
            summary = article.summary
            m = re.match(find, summary)
            print(i, med_terms[i], '\n', m.group(1))
            print('\n')
            explanation = m.group(1)
            explanation_list.append(explanation)
#             wiki_dict[med_terms[i]] = m.group(1)
#             print(med_terms[i], wiki_dict[med_terms[i]])

#     wiki_list.append(wiki_dict)
```
